### 1. Write a python program that will calculate the function sin(x). Make sure your program
meets the following requirements:
1. The value of x can be specified and changed in the code (or as user input).
2. The value of x0 can be specified and changed in the code (or as user input).
3. The definition of fractional error is:
ϵ=|

true |
This can be expressed as percent error by multiplying the above equation by
100%. Note ftrue is the true/known value of f and fapprx is the approximate value
from the Taylor series expansion with N terms.
Your program should continue to calculate the Talyor series until ϵ≤10−6
.
4. Demonstrate your program works correctly by running it for the following case: x0 =
p/12, x = p/2.

```python
import math
from tabulate import tabulate

def fprime(x, k):
    if k == 0:
        return math.sin(x)
    if k == 1:
        return math.cos(x)
    if k == 2:
        return -math.sin(x)
    if k == 3:
        return -math.cos(x)

def mainUser():
    x0 = float(input("x0 = "))  
    x = float(input("x = "))
    h = x - x0
    fapprx = 0.0
    ftrue = math.sin(x)
    err_stop = 1.0 * (10 ** -6)
    true_err = 1.1 * err_stop
    max_iter = 100
    f_string = "f"
    i_string = "i"
    true_err_string = "true err"
    table = [[i_string, f_string, true_err_string]]
    for i in range(0, max_iter):
        k = (i + 4) % 4
        fapprx += fprime(x0, k) * h ** i / math.factorial(i)
        true_err = abs((ftrue - fapprx) / ftrue)
        if true_err <= err_stop:
            table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
            break
        table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
    print(tabulate(table, tablefmt="fancy_grid", headers="firstrow"))
                
def mainConstant():
    x0 = math.pi/12  
    x = math.pi/2
    h = x - x0
    fapprx = 0.0
    ftrue = math.sin(x)
    err_stop = 1.0 * (10 ** -6)
    true_err = 1.1 * err_stop
    max_iter = 100
    f_string = "f"
    i_string = "i"
    true_err_string = "true err"
    table = [[i_string, f_string, true_err_string]]
    for i in range(0, max_iter):
        k = (i + 4) % 4
        fapprx += fprime(x0, k) * h ** i / math.factorial(i)
        true_err = abs((ftrue - fapprx) / ftrue)
        if true_err <= err_stop:
            table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
            break
        table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
    print(tabulate(table, tablefmt="fancy_grid", headers="firstrow"))
        
if input("Use your own input y/n: ") == "y":
    mainUser()
else:
    print("Using constant x0 = pi/12 and x = pi/2")
    mainConstant()
```

### 3. Use your program from #4 to use base points, x0 = 0.0, 0.5, 1.5, 2.0, 2.5, 3.0, 3.5 and x = 4.0 to determine the number of terms, Nϵ required to get an answer that has ϵ≤10^−10
1. Make a graph of Nϵ versus h (Note h = x-x0)

```python
import math
from tabulate import tabulate
import matplotlib.pyplot as plt
 
def fprime(x, k):
    if k == 0:
        return math.exp(x)
    if k == 1:
        return math.exp(x)
    if k == 2:
        return math.exp(x)
    if k == 3:
        return math.exp(x)                
def mainConstant():
    list1 = [0.0,0.5,1.5,2.0,2.5,3.0,3.5]
    plotx = []
    ploty = [] 
    for index in range(0,len(list1)):
        x0 = list1[index] 
        x = 4.0
        h = x - x0
        print(f"\nUsing constant x0 = {x0} and x = {x}")
        fapprx = 0.0
        ftrue = math.exp(x)
        err_stop = 1.0 * (10 ** -10)
        true_err = 1.1 * err_stop
        max_iter = 100
        f_string = "f"
        i_string = "i"
        true_err_string = "true err"
        table = [[i_string, f_string, true_err_string]]
        for i in range(0, max_iter):
            k = (i + 4) % 4
            fapprx += fprime(x0, k) * h ** i / math.factorial(i)
            true_err = abs((ftrue - fapprx) / ftrue)
            if true_err <= err_stop:
                table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
                break
            table.append([i + 1, f"{fapprx:.15e}", f"{true_err:.2e}"])
        print(tabulate(table, tablefmt="fancy_grid", headers="firstrow"))
        plotx.append(i+1)
        ploty.append(h)    
    plt.plot(plotx, ploty)
    plt.xlabel('Ne - axis')
    plt.ylabel('h - axis')
    plt.title('Ne vs. h plot')
    plt.scatter(plotx, ploty, s=80, facecolors='none', edgecolors='r')
    plt.show()
        
mainConstant()
```
